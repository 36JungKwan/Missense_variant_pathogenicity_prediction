```python
# Danh sách giá trị cần kiểm tra
invalid_values = ["-", "na", "n/a", "NA", "N/A", "NaN", "nan"]

# Thay thế toàn bộ giá trị lỗi trong toàn DataFrame bằng np.nan
df_official = df_official.replace(invalid_values, np.nan)
```

```python
df_official.isnull().sum()
```

```python
df_official = df_official.rename(columns={"Chromosome": "CHROM"})
df_official = df_official.rename(columns={"PositionVCF": "POS"})
df_official = df_official.rename(columns={"ReferenceAlleleVCF": "REF"})
df_official = df_official.rename(columns={"AlternateAlleleVCF": "ALT"})
def normalize_chromosome(chrom_name):
    chrom_str = str(chrom_name).strip() # Chuyển sang string và xóa khoảng trắng

# Lọc bỏ các contig không xác định (Unplaced/random/alt)
    if 'Un' in chrom_str or 'random' in chrom_str or 'alt' in chrom_str:
        return None # Sẽ bị xóa ở bước tiếp theo

    # Xử lý các trường hợp đặc biệt (MT -> chrM)
    if chrom_str.upper() == 'MT':
        return 'chrM'

    # Nếu nó đã có 'chr' rồi thì thôi (như chr1, chrX, chrM của bạn)
    if chrom_str.startswith('chr'):
        return chrom_str

    # Nếu nó là số (1, 2, X, Y) thì thêm 'chr'
    return 'chr' + chrom_str

print("Đang chuẩn hóa tên chromosome ...")
df_official['CHROM'] = df_official['CHROM'].apply(normalize_chromosome)

original_count = len(df_official)
```

```python
df_official.isnull().sum()
```

```python
df_official = df_official.dropna(subset=['CHROM']) # Xóa các hàng (như chrUn) đã bị gán None
df_official = df_official.dropna(subset=['POS'])
df_official = df_official.dropna(subset=['REF'])
df_official = df_official.dropna(subset=['ALT'])
removed_count = original_count - len(df_official)
print(f"Đã lọc bỏ {removed_count} hàng ('chrUn' hoặc contig rác, thiếu POS, ALT, REF).")

print("Đang sắp xếp df_official (theo chrom và pos)...")
# 1. Đảm bảo 'pos' là kiểu số (integer) để sắp xếp đúng
df_official['POS'] = pd.to_numeric(df_official['POS'])
# 2. Sắp xếp theo 'chrom' trước, sau đó theo 'pos'
df_official = df_official.sort_values(by=["CHROM", "POS"])
df_official
```
